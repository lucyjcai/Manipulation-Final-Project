```python
import trimesh
import numpy as np
from pathlib import Path

# Load the mesh (it handles the parsing)

def fix_normals(input_obj_path, output_obj_path, rewrite = False):
    if input_obj_path == output_obj_path and not rewrite:
        raise ValueError("Input and output paths are the same. To overwrite, set rewrite=True.")
    
    mesh = trimesh.load(input_obj_path)

    print(f"Loaded mesh with {len(mesh.vertices)} vertices and {len(mesh.faces)} faces.")
    mesh.fix_normals()
    mesh.export(output_obj_path, file_type='obj', include_normals=True)

    print(f"Success! Saved {output_obj_path} with valid normals.")
```

```python
input_obj_path = f"{Path.cwd()}/urdf/custom/meshes/parts/drawer.obj"
```

```python
fix_normals(input_obj_path, input_obj_path, rewrite=True)
```
